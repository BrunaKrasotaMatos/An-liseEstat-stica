```r
titanic_test <- read.csv(file = 'titanic_dw_test.csv')
titanic_train <- read.csv(file = 'titanic_dw_train.csv')
```

```r
titanic_test$Sex <- as.ordered(gsub(",","",titanic_test$Sex))
titanic_test$Cabin <- as.ordered(gsub(",","",titanic_test$Cabin))
titanic_test$Embarked <- as.ordered(gsub(",","",titanic_test$Embarked))

head(titanic_test)
summary(titanic_test)
```

```r

titanic_train$Sex <- as.ordered(gsub(",","",titanic_train$Sex))
titanic_train$Cabin <- as.ordered(gsub(",","",titanic_train$Cabin))
titanic_train$Embarked <- as.ordered(gsub(",","",titanic_train$Embarked))

head(titanic_train)
summary(titanic_train)
```

```r
# selecionando apenas as colunas solicitadas
# Survived, Pclass, Sex, Age, SibSp, Parch, Fare, Embarked
train_frame = data.frame(titanic_train[,c("Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked","PassengerId")])
head(train_frame)
```

```r
install.packages("AER")
```

```r
data(train_frame, package = "AER")
summary(train_frame)
```

```r
# Você deve  atribuir um valor para os NAs na variável Age.
#str(train_frame, strict.width = "wrap")
media_age = mean(train_frame$Age)
media_age
media_age_se_na = mean(train_frame$Age, na.rm = TRUE)
media_age_se_na
# Estou com elementeo NA que está dando problema no meu valor de media
train_frame$Age[is.na(train_frame$Age)]<-media_age_se_na
head(train_frame$Age)
```

```r
# Você deve remover as linhas onde ainda estiverem faltando dados
train_frame_2 = na.omit(train_frame)
str(train_frame_2)
```

```r
# Você deve criar um modelo  onde Survived será uma função das demais variáveis.
modelo_surv = glm(Survived ~ PassengerId + Pclass + Sex + Age + SibSp + Parch + Fare + Embarked, family = binomial, data = train_frame_2)
summary(modelo_surv)
plot(modelo_surv)
```
